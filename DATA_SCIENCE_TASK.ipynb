```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import f1_score


d12['plot_keywords_joined'] = d12['plot_keywords'].apply(lambda x: ' '.join(x))


label_encoder = LabelEncoder()
d12['director_name_encoded'] = label_encoder.fit_transform(d12['director_name'])


mlb = MultiLabelBinarizer()
y = mlb.fit_transform(d12['genres'])

# Create TF-IDF features for the 'plot_keywords_joined'
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=10000)
X_tfidf = tfidf_vectorizer.fit_transform(d12['plot_keywords_joined'])

# Combine 'director_name_encoded' with 'plot_keywords' tfidf features
import scipy.sparse as sp
X_combined_features = sp.hstack((X_tfidf, d12[['director_name_encoded']].values), format='csr')

# Split the dataset into training and validation set
X_train, X_test, y_train, y_test = train_test_split(X_combined_features, y, test_size=0.2, random_state=42)

# Train the OneVsRest multi-label classifier with Logistic Regression
lr = LogisticRegression(solver='lbfgs')
clf = OneVsRestClassifier(lr)
clf.fit(X_train, y_train)

# Predict on the validation set
y_pred = clf.predict(X_test)

# Evaluate the model
f1 = f1_score(y_test, y_pred, average="micro")
print(f'F1 score: {f1}')

# Inference function for predicting genres
def infer_tags(director_name, plot_keywords):
    # Transform the director's name using the label encoder
    director_features = label_encoder.transform([director_name])

    # Ensure plot_keywords is a string
    plot_keywords = ' '.join(plot_keywords) if isinstance(plot_keywords, list) else plot_keywords

    # Transform the plot_keywords to tfidf features
    plot_keywords_features = tfidf_vectorizer.transform([plot_keywords])

    # Combine director_features and plot_keywords_features for the final feature vector
    features_combined = sp.hstack((plot_keywords_features, director_features.reshape(-1, 1)), format='csr')

    # Predict genres
    predicted_genres_encoded = clf.predict(features_combined)

    # Inverse transform to get the genre names
    predicted_genres = mlb.inverse_transform(predicted_genres_encoded)

    return predicted_genres

# Example usage of the inference function
example_director = 'Edward Burns'
example_keywords = d12.loc[d12['director_name'] == example_director, 'plot_keywords'].values[0]
predicted_genres = infer_tags(example_director, example_keywords)

print(f'Predicted genres for director {example_director}: {predicted_genres}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import f1_score, accuracy_score, mean_squared_error

# Load your dataframe here
# d3 = pd.read_csv('path_to_your_data.csv')

# Ensure proper data types and drop NaNs in 'title_year'
d3['title_year'] = pd.to_numeric(d3['title_year'], errors='coerce')
d3.dropna(subset=['title_year'], inplace=True)
d3['title_year'] = d3['title_year'].astype(int)

# Sort by director name and title year for correct gap calculation
d3.sort_values(by=['director_name', 'title_year'], inplace=True)
d3['previous_release_year'] = d3.groupby('director_name')['title_year'].shift(1)
d3['time_between_films'] = d3['title_year'] - d3['previous_release_year']

# Apply a cap to 'time_between_films' at 10 years and ensure no negative values
d3['time_between_films'] = d3['time_between_films'].clip(lower=0, upper=10)

# Calculate the average gap, taking care of NaNs
d3['avg_gap'] = d3.groupby('director_name')['time_between_films'].transform('mean').fillna(0).clip(lower=0)

# Feature Selection and Preprocessing for Genres Prediction
mlb = MultiLabelBinarizer()
genres_encoded = mlb.fit_transform(d3['genres'])
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=10000)
plot_keywords_tfidf = tfidf_vectorizer.fit_transform(d3['plot_keywords'].astype('str'))

# Encode the director's name
label_enc = LabelEncoder()
directors_encoded = label_enc.fit_transform(d3['director_name'])

# Prepare feature and target matrices for modeling
X_genres = np.hstack((plot_keywords_tfidf.toarray(), directors_encoded.reshape(-1, 1)))
y_genres = genres_encoded
X_release_year = d3[['imdb_score', 'duration', 'avg_gap']].values
y_release_year = d3['title_year'] + d3['avg_gap']

# Split the datasets
X_train_genres, X_test_genres, y_train_genres, y_test_genres = train_test_split(X_genres, y_genres, test_size=0.2, random_state=42)
X_train_year, X_test_year, y_train_year, y_test_year = train_test_split(X_release_year, y_release_year, test_size=0.2, random_state=42)

# Modeling and Training
genres_clf = MultiOutputClassifier(RandomForestClassifier(random_state=42))
genres_clf.fit(X_train_genres, y_train_genres)

release_year_clf = RandomForestRegressor(random_state=42)
release_year_clf.fit(X_train_year, y_train_year)

# Predictions and Evaluation
y_pred_genres = genres_clf.predict(X_test_genres)
f1 = f1_score(y_test_genres, y_pred_genres, average='micro')
accuracy = accuracy_score(y_test_genres, y_pred_genres)
rmse = mean_squared_error(y_test_year, y_pred_year, squared=False)

print(f'F1 score for genres prediction: {f1:.2f}')
print(f'RMSE for release year prediction: {rmse:.2f}')

# Function to make predictions and format the output
def predict_and_format(director_name):
    try:
        director_index = label_enc.transform([director_name])[0]
        plot_keywords_vector = tfidf_vectorizer.transform([''])  # Assuming no plot keywords for this prediction
        features = np.hstack((plot_keywords_vector.toarray(), [[director_index]]))
        genres_encoded = genres_clf.predict(features)
        genres = mlb.inverse_transform(genres_encoded)[0]

        director_features = X_release_year[directors_encoded == director_index].mean(axis=0).reshape(1, -1)
        predicted_year = release_year_clf.predict(director_features)[0]
        next_year = int(np.ceil(predicted_year))
    except ValueError:
        avg_gap = d3['avg_gap'].mean()
        last_year = d3['title_year'].max()
        next_year = int(last_year + avg_gap)
        genres = ['Unknown']  # Default genres if the director is unknown
    return {
        'Name of Director': director_name,
        'Next movie release': next_year,
        'Genres': genres
    }

# Example usage
example1 = predict_and_format('Barry Sonnenfeld')
example2 = predict_and_format('Zack Snyder')

# Print the examples
print('Example1:', example1)
print('Example2:', example2)
```
